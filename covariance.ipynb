```python
import regimes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1)
```

```python
# Normalization (Standard)
def normalize(data, type='minmax'):

    if type=='std':
        return (np.array(data) - np.mean(data))/np.std(data)
        
    elif type=='minmax':
        return (np.array(data) - np.min(data))/(np.max(data) - np.min(data))
```

```python
#reading of the data for each site
DE_Hai=pd.read_csv(r'../datasets/Subsets/FLX_DE-Hai_FLUXNET2015_SUBSET_DD_2000-2012_1-4.csv')
IT_Col=pd.read_csv(r'../datasets/Subsets/FLX_IT-Col_FLUXNET2015_SUBSET_DD_1996-2014_1-4.csv')

df1 = DE_Hai[["TA_F","NEE_VUT_REF","RECO_NT_VUT_REF","GPP_NT_VUT_REF","RECO_DT_VUT_REF","GPP_DT_VUT_REF"]]
df2=IT_Col[["TA_F","NEE_VUT_REF","RECO_NT_VUT_REF","GPP_NT_VUT_REF","RECO_DT_VUT_REF","GPP_DT_VUT_REF"]]
```

```python
#normalization of data

df1=df1.apply(normalize)
df2=df2.apply(normalize)
```
